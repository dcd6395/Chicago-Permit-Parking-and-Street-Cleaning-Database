```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "./Resources/Parking_Permit_Zones.csv"
parking_df = pd.read_csv(csv_file)
parking_df.head()
```

```python
parking_clean = parking_df.drop(columns='SECOND STREET DIRECTION')
```

```python
parking_clean.head()
```

```python
parking_renamed_df = parking_clean.rename(columns={"ROW ID": "id",
                                      "STATUS": "status",
                                      "ZONE": "zone",
                                      "ODD_EVEN": "odd_even",
                                        "ADDRESS RANGE - LOW": "address_range_low",
                                        "ADDRESS RANGE - HIGH": "address_range_high",
                                        "STREET DIRECTION": "street_direction",
                                        "STREET NAME": "street_name",
                                        "STREET TYPE": "street_type",
                                        "BUFFER" : "buffer",
                                        "WARD - LOW": "ward",
                                        "WARD - HIGH": "ward_high"})

parking_renamed_df.head()
```

```python
parking_renamed_df=parking_renamed_df.dropna()
parking_renamed_df
```
